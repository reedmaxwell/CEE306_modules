### Width,Q

```python
%matplotlib inline
fig = plt.figure()

# linear
xmin=150000
xmax=2500000
ymin=1700
ymax=3000
ax1 = df.plot.scatter(x='Total_Q_cfs',y='Width_ft',c='DarkBlue',figsize=(10,6))
ax1.set_xlim(xmin,xmax)
#ax1.set_ylim(ymin,ymax)
ax1.grid(True)
ax1.set_ylabel("Width, ft")
ax1.set_xlabel("Discharge, cfs")


# ln
lwidth = np.log(df['Width_ft'])
df1=lwidth.to_frame()
lq = np.log(df['Total_Q_cfs'])
df2=lq.to_frame()
df_ln = pd.merge(df1,df2,left_index=True,right_index=True)

xmin=12
xmax=15
ymin=7.4
ymax=8.6
ax2 = df_ln.plot.scatter(x='Total_Q_cfs',y='Width_ft',c='DarkBlue',figsize=(10,6))
ax2.set_ylim(ymin,ymax)
ax2.set_xlim(xmin,xmax)
ax2.grid(True)
ax2.set_ylabel("ln(Width), ft")
ax2.set_xlabel("ln(Discharge), cfs")
```

### Depth,Q

```python
%matplotlib inline
df['Dmean'] = df['Area_sqft']/df['Width_ft']
fig = plt.figure()

# linear
xmin=150000
xmax=2500000
ymin=30
ymax=80
ax1 = df.plot.scatter(x='Total_Q_cfs',y='Dmean',c='DarkBlue',figsize=(10,6))
ax1.set_xlim(xmin,xmax)
ax1.grid(True)
ax1.set_xlim(xmin,xmax)
ax1.set_ylim(ymin,ymax)
ax1.set_ylabel("Mean Depth, ft")
ax1.set_xlabel("Discharge, cfs")

# ln
lDmean = np.log(df['Dmean'])
df1=lDmean.to_frame()
lq = np.log(df['Total_Q_cfs'])
df2=lq.to_frame()
df_ln = pd.merge(df1,df2,left_index=True,right_index=True)

xmin=12
xmax=15
ymin=3.5
ymax=4.5
ax2 = df_ln.plot.scatter(x='Total_Q_cfs',y='Dmean',c='DarkBlue',figsize=(10,6))
ax2.set_xlim(xmin,xmax)
ax2.set_ylim(ymin,ymax)
ax2.grid(True)
ax2.set_ylabel("ln(Mean Depth), ft")
ax2.set_xlabel("ln(Discharge), cfs")
```
